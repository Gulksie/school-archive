# 3 Decision Stuctures
Decision structures allow programs to perform specific tasks based on the predefined conditions.

# 3.1 The If Statement
The `if` statement is used to test a conditional statements and execute a block of code if the condition is found to be `True`. <br/>
An If Statement in Python is defined by:
```
if <condition statment>:
    <statement 1>
    <statement 2>
    <...>
    <statement n>
```

* When the conditional evaluates to `True`, the indented block of code is executed. 
* When the conditional evaluates to `False`, the indented block of code is skipped. 

````
# IMPLEMENT AN "if" STATEMENT IN THIS CELL 
game = input('What is the best video game ever?')

if game == "Tetris":
    print("Yes, Tetris is the best video game of all time")
````

# 3.2 The If-Else Statement
The `else` keyword catches any condition which isn't caught by the preceding conditions in the if statement. <br/>
* When the conditional in the `if` statement is `False`, the code block under the `else` statement is executed.
* When the conditional in the `if` statement is `True`, the code block under the `else` statement is skipped.

<br/>
An If-Else Statement in Python is defined by:

```
if <condition>:
    <statement 1>
    <statement 2>
    <...>
    <statement n>

else:
    <statement 1>
    <statement 2>
    <...>
    <statement n>
```

````
# IMPLEMENT AN "if-else" STATEMENT IN THIS CELL 
age = 18

if age >= 18:
    print("You are an adult, make sure to vote!")
  
# ADD AN ELSE STATEMENT HERE
````
